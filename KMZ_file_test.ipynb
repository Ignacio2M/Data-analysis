```python
import zipfile
from pykml import parser
```

```python
with zipfile.ZipFile('Data/mapa-contenedores-papelaraba2019.kmz',     "r") as kmz:
    kmz.extractall("Data/mapa-contenedores-papelaraba2019")
```

```python
def create_papelera(id, point, values):
    name = ["cuadrilla", "municipio", 
            "localidad", "cod_cuadri", "cod_munici", 
            "cod_locali", "dirección", "x_etrs89", 
            "y_etrs89", "color", "modelo", "adaptado", 
            "volumen", "estado_gen", "serigrafia", "tapas", 
            "vaso", "compuertas", "limpieza", "usuarios", 
            "entorno", "frec_invi", "dias_invie", "dias_veran", 
            "observacio"]
    data = dict(zip(name, values))
    point = str(point).split(',')
    point = [float(numero) for numero in point]
    data.update({"id":id, "point": point})
    return data
```

```python
with open("Data/mapa-contenedores-papelaraba2019/doc.kml") as f:
  doc = parser.parse(f).getroot()
```

```python
f = doc.Folder
```

```python
d = f.getchildren()[2]
```

```python
p = d.Folder.getchildren()[1]
```

```python
data = []
for p in d.Folder.getchildren():
    try:
        data.append(create_papelera(p.name, p.Point.coordinates, p.ExtendedData.SchemaData.getchildren()))
    except:
        pass
```
